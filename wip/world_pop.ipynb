```python
import pandas as pd
```

```python
world = pd.read_csv('/Users/vino/Desktop/project3copy/proj_3/population.csv')
world
```

```python


# Set the correct header (first row after skipping should be used as the header)
world.columns = world.iloc[0]  # Set the first row as the header
world = world[1:]  # Remove the first row after setting it as header

# If there are unnamed columns or unnecessary columns, drop them
world = world.drop(columns=[col for col in world.columns if 'Unnamed' in col])

# Reset the index to remove any unnecessary columns from being part of the index
world.reset_index(drop=True, inplace=True)

# Now you should have a DataFrame with the correct headers and clean data
print(world.head())
```

```python
world.head()
```

```python
world.head()
world.dropna()
world
world = world[['Name', 'Total Population','Year']].copy()
world
```

```python
world = world.drop(world.index[0])
world
```

```python

# Assuming merged_data is your main dataframe

# Clean the 'Total Population' column by removing commas and spaces
world['Total Population'] = world['Total Population'].str.replace(',', '').str.strip()

# Convert the cleaned 'Total Population' column to numeric
world['Total Population'] = pd.to_numeric(world['Total Population'], errors='coerce')

# Verify that the 'Total Population' column is now numeric
print(world[['Total Population']].dtypes)
print(world[['Total Population']].head())

# Check for any missing values that might have been introduced during the cleaning process
missing_values = world['Total Population'].isna().sum()
print(f"Number of missing values in 'Total Population': {missing_values}")
```

```python
# Option 2: Fill missing values with the mean of the column
world['Total Population'] = world['Total Population'].fillna(world['Total Population'].mean())
```

```python
print(world[['Total Population']].isna().sum())
print(world[['Total Population']].head())
```
