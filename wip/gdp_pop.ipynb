```python
import pandas as pd

# Load the datasets
gdp_df = pd.read_csv('proj 3/world_gdp1960.csv')
pop_df = pd.read_csv('proj 3/world_pop_1950.csv')

# Inspect the first few rows
print(gdp_df.head())
print(pop_df.head())

# Merge the datasets on the 'Country'/'Name' columns
merged_df = pd.merge(gdp_df, pop_df, left_on='Country', right_on='Name', how='inner')

# Display the merged dataframe
print(merged_df.head())
```
